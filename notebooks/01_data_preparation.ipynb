```python
# List of carbapenem antibiotics of interest
carbapenems = ['imipenem', 'meropenem', 'doripenem', 'ertapenem']

# -------------------------------------------------------------------
# Filter AMR records for carbapenem-related antibiotics
# Using `.copy()` avoids chained assignment issues downstream
# -------------------------------------------------------------------
carbapenem_data = (
    amr_records.loc[
        amr_records['Antibiotic'].str.endswith('enem', na=False)
    ].copy()
)

# -------------------------------------------------------------------
# Identify genomes originating from India
# `na=False` prevents errors if location contains missing values
# -------------------------------------------------------------------
indian_genome_records = (
    genome_records.loc[
        genome_records['Geographic Location'].str.startswith('India', na=False)
    ].copy()
)

# -------------------------------------------------------------------
# Summary statistics
# -------------------------------------------------------------------
n_indian_genomes = indian_genome_records['Genome ID'].nunique()
n_carbapenem_genomes = carbapenem_data['Genome ID'].nunique()

print(f'Indian genomes: {n_indian_genomes}')
print(f'Carbapenem-associated genomes: {n_carbapenem_genomes}')

# -------------------------------------------------------------------
# Exclude Indian genomes from training and test dataset
# These genomes will be used post the model is build (production phase)
# -------------------------------------------------------------------
indian_genome_ids = set(indian_genome_records['Genome ID'])

indian_carbapenem_data = carbapenem_data[carbapenem_data['Genome ID'].isin(indian_genome_ids)]

carbapenem_data = (
    carbapenem_data.loc[
        ~carbapenem_data['Genome ID'].isin(indian_genome_ids)
    ].copy()
)
```

```python
# Build the summary
rows = []
for ab in carbapenems:
    subset = carbapenem_data[carbapenem_data['Antibiotic'] == ab]
    sign_counts = subset['Measurement Sign'].value_counts()
    rows.append({
        'Antibiotic':       ab,
        '= (exact)':        sign_counts.get('=',  0),
        '<= (left-cens.)':  sign_counts.get('<=', 0),
        '> (right-cens.)':  sign_counts.get('>',  0),
        '>= (right-cens.)': sign_counts.get('>=', 0),
        'Total Records':    len(subset),
        'Unique Genome IDs': subset['Genome ID'].nunique()
    })

summary_df = pd.DataFrame(rows).set_index('Antibiotic')
```

```python
# Exclude carbapenem (class-level, no drug specificity)
# Focus on records WITH a Resistant Phenotype label

for ab in ['imipenem', 'meropenem', 'doripenem', 'ertapenem']:
    subset = carbapenem_data[carbapenem_data['Antibiotic'] == ab]
    
    has_phenotype = subset['Resistant Phenotype'].notna().sum()
    has_mic       = subset['Measurement Sign'].notna().sum()
    phenotype_dist = subset['Resistant Phenotype'].value_counts().to_dict()
    
    print(f"\n{ab.upper()}")
    print(f"  Records with phenotype label : {has_phenotype}")
    print(f"  Records with MIC measurement : {has_mic}")
    print(f"  Phenotype distribution       : {phenotype_dist}")
    print(f"  Unique Genome IDs            : {subset['Genome ID'].nunique()}")
```

```python
# =============================================================================
# COLLAPSE INTERMEDIATE (I) + RESISTANT (R) → NON-SUSCEPTIBLE (NS)
# Terminology follows Hicks et al. (2019), which explicitly defines
# Non-Susceptibility (NS) as the combined I + R category for clinical
# AMR binary classification — used in their A. baumannii analysis.
#
# Clinical justification: For carbapenems, both Intermediate and Resistant
# MICs exceed achievable serum concentrations under standard dosing regimens.
# The clinical question is not "is this resistant?" but "is this susceptible?"
# Any organism that is not clearly Susceptible is Non-Susceptible (NS).
# =============================================================================

# Define target antibiotics (ertapenem excluded — insufficient labeled data)
target_carbapenems = ['imipenem', 'meropenem', 'doripenem']

# Work on a copy to preserve the original carbapenem_data intact
modeling_data = carbapenem_data[
    carbapenem_data['Antibiotic'].isin(target_carbapenems) &
    carbapenem_data['Resistant Phenotype'].notna()
].copy()

# -----------------------------------------------------------------------------
# Collapse: Intermediate → Non-Susceptible
#           Resistant    → Non-Susceptible
#           Susceptible  → Susceptible (unchanged)
#
# Explicit mapping of all three source classes avoids silent passthrough
# and makes the collapse intent unambiguous to any reader of the code.
# -----------------------------------------------------------------------------
phenotype_map = {
    'Resistant'    : 'Non-Susceptible',
    'Intermediate' : 'Non-Susceptible',
    'Susceptible'  : 'Susceptible'
}

modeling_data['Binary Label'] = modeling_data['Resistant Phenotype'].map(phenotype_map)

# Confirm only two classes remain
print("Unique labels after collapse:")
print(modeling_data['Binary Label'].unique())

# -----------------------------------------------------------------------------
# Verify no unmapped values slipped through (would appear as NaN)
# -----------------------------------------------------------------------------
unmapped = modeling_data['Binary Label'].isna().sum()
assert unmapped == 0, (
    f"{unmapped} records could not be mapped — "
    f"check for unexpected Resistant Phenotype values: "
    f"{modeling_data.loc[modeling_data['Binary Label'].isna(), 'Resistant Phenotype'].unique()}"
)
print("Assertion passed: all records successfully mapped to S / NS.\n")

# -----------------------------------------------------------------------------
# Per-antibiotic class distribution after collapse
# -----------------------------------------------------------------------------
print("Class distribution after I + R → Non-Susceptible collapse:\n")

summary_rows = []

for ab in target_carbapenems:
    subset = modeling_data[modeling_data['Antibiotic'] == ab]
    counts = subset['Binary Label'].value_counts()
    n_total          = len(subset)
    n_susceptible    = counts.get('Susceptible',     0)
    n_nonsusceptible = counts.get('Non-Susceptible', 0)
    pct_s  = round(100 * n_susceptible    / n_total, 1)
    pct_ns = round(100 * n_nonsusceptible / n_total, 1)

    # Breakdown of what went into NS (for traceability)
    orig_counts = subset['Resistant Phenotype'].value_counts()
    n_r = orig_counts.get('Resistant',    0)
    n_i = orig_counts.get('Intermediate', 0)

    summary_rows.append({
        'Antibiotic'              : ab,
        'Susceptible (n)'         : n_susceptible,
        'Non-Susceptible (n)'     : n_nonsusceptible,
        '  of which Resistant'    : n_r,
        '  of which Intermediate' : n_i,
        'Total (n)'               : n_total,
        'Susceptible (%)'         : pct_s,
        'Non-Susceptible (%)'     : pct_ns,
        'Unique Genome IDs'       : subset['Genome ID'].nunique()
    })

    print(f"{ab.upper()}")
    print(f"  Susceptible      : {n_susceptible:>4}  ({pct_s}%)")
    print(f"  Non-Susceptible  : {n_nonsusceptible:>4}  ({pct_ns}%)")
    print(f"    ↳ Resistant    : {n_r:>4}")
    print(f"    ↳ Intermediate : {n_i:>4}")
    print(f"  Total            : {n_total:>4}")
    print(f"  Unique Genomes   : {subset['Genome ID'].nunique()}\n")

summary_collapse_df = pd.DataFrame(summary_rows).set_index('Antibiotic')
print(summary_collapse_df)

# -----------------------------------------------------------------------------
# Sanity check: confirm Intermediate has been fully absorbed into NS
# -----------------------------------------------------------------------------
residual_intermediate = (modeling_data['Binary Label'] == 'Intermediate').sum()
assert residual_intermediate == 0, (
    "Intermediate records still present — collapse failed."
)
print("\nAssertion passed: no Intermediate records remain in Binary Label.")
```

```python
selected_columns = ['Genome ID', 'Genome Name', 'Antibiotic', 
                    'Measurement', 'Measurement Sign', 'Measurement Value', 
                    'Evidence', 'Resistant Phenotype', 'Binary Label']

modeling_data = modeling_data[selected_columns].reset_index(drop=True)
```
